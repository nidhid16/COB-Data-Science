```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("C:\\Users\\Nidhi Davane\\Documents\\dataset - netflix1 (1).csv")
```

```python
df.head()
```

```python
plt.figure(figsize=(10,5))
sns.histplot(df["release_year"], bins=30, kde=True)
plt.title("Distribution of the release year")
plt.xlabel("Release Year")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(x="type", data=df)
plt.title("Distribution of the show types")
plt.xlabel("Show Type")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(x="country", data=df, order=df["country"].value_counts().index[:10])
plt.title("Top 10 countries with the most shows")
plt.xlabel("Country")
plt.ylabel("Count")
plt.xticks(rotation=45, ha="right")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(x="rating", data=df, order=df["rating"].value_counts().index)
plt.title("Distribution of Ratings")
plt.xlabel("Rating")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.histplot(df[df["type"] == "Movie"]["duration"], bins=30, kde=True)
plt.title("Distribution of Movie Duration")
plt.xlabel("Duration (minutes)")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(10, 5))
tv_show_genres = df[df["type"] == "TV Show"]["listed_in"].str.split(", ", expand=True).stack().value_counts().head(10)
sns.barplot(x=tv_show_genres.values, y=tv_show_genres.index, orient="h")
plt.title("Top 10 Genres for TV Shows")
plt.xlabel("Count")
plt.ylabel("Genre")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.histplot(x="release_year", y="rating", data=df, alpha=0.5)
plt.title("Release Year vs Rating")
plt.xlabel("Release Year")
plt.ylabel("Rating")
plt.show()
```
